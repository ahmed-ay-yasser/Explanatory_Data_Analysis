### <span style="color:#00008b"> 2. Bivariate Analysis </span>

```python
#Sales Orders by Ship Mode
df.groupby("Ship Mode")[["Sales", "Cost", "Profit"]].sum().sort_values(by=["Sales"], ascending = False).round()
```

```python
#Sales Orders by Region
df.groupby("Region")[["Sales", "Cost", "Profit"]].sum().sort_values(by=["Sales"], ascending = False).round()
```

```python
#Sales Orders by Categories
df.groupby("Category")[["Sales", "Cost", "Profit"]].sum().sort_values(by=["Sales"], ascending = False).round()
```

```python
#Sales Orders by Sub-Categories
df.groupby("Sub-Category")[["Sales", "Profit"]].sum().sort_values(by = "Profit").round()
```

```python
#Exploring top 10 state in sales
df.groupby("State")[["Sales", "Profit"]].sum().sort_values(by = ["Sales"], ascending  = False).round().head(10)
```

```python
#Exploring top 10 states achieved net loss
Loss_df = df[df["Cost"] > df["Sales"]]
```

```python
Loss_df.groupby("State")[["Cost", "Sales"]].sum().sort_values(by = ["Cost"], ascending = False).round().head(10)
```
